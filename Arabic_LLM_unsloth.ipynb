```python
import pandas as pd

data_path = "/content/drive/MyDrive/MentalQA/Train_Dev.tsv"
df = pd.read_csv(data_path, sep="\t")
df.head()
```

```python
from datasets import Dataset

# Combine question & answer into prompt → response format
dataset = Dataset.from_pandas(df[["question", "answer"]].rename(
    columns={"question": "instruction", "answer": "output"}
))

def format_example(example):
    return f"""### Instruction:
{example['instruction']}

### Response:
{example['output']}"""

dataset = dataset.map(lambda x: {"text": format_example(x)})
```

```python
from datasets import Dataset

# Combine question and answer into one prompt
df["text"] = df.apply(lambda row: f"سؤال: {row['question']}\nإجابة: {row['answer']}", axis=1)

dataset = Dataset.from_pandas(df)
```

```python
from unsloth import FastLanguageModel
import torch

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Qwen3-1.7B-unsloth-bnb-4bit",
    max_seq_length = 2048,
    dtype = torch.float16,
    load_in_4bit = True,
)

model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    lora_alpha = 32,
    lora_dropout = 0.05,
    bias = "none",
)

```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir = "qwen3-mentalqa-lora",
    per_device_train_batch_size = 1,
    gradient_accumulation_steps = 4,
    warmup_steps = 5,
    num_train_epochs = 3,
    learning_rate = 2e-4,
    logging_steps = 5,
    fp16 = True,
    lr_scheduler_type = "cosine",
    optim = "paged_adamw_8bit",
    save_total_limit = 1,
)
```

```python
from trl import SFTTrainer

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = dataset.select(range(10)),  # small eval for test
    dataset_text_field = "text",
    max_seq_length = 512,
    args = training_args,
)

trainer.train()
```

```python
# Set save path inside your Drive
save_path = "/content/drive/MyDrive/MentalQA/qwen3-mentalqa-model"

# Save model & tokenizer
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)
from unsloth import FastLanguageModel
FastLanguageModel.for_inference(model)  # unpatch

# Inference function
def generate(text):
    prompt = f"### Instruction:\n{text}\n\n### Response:"
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=100)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Test
generate("ما هي أعراض القلق المزمن؟")
```
